```python
# Remove rows where all elements are NaN
df_cleaned = df.dropna(how='all')
display(df_cleaned.head())
```

```python
new_column_names = [
    'State_Abbreviation',
    'Institution_Name',
    'Total_Enrollment',
    'Enrollment_In_State_Percent',
    'Enrollment_Out_of_State_Percent',
    'Enrollment_Men_Percent',
    'Enrollment_Women_Percent',
    'Total_Degrees_Awarded',
    'Degrees_In_State_Percent',
    'Degrees_Out_of_State_Percent',
    'Degrees_Men_Percent',
    'Degrees_Women_Percent'
]

df_cleaned.columns = new_column_names
display(df_cleaned.head())
```

```python
def flatten_multiindex_columns(df):
    new_columns = []
    for col in df.columns:
        # Join non-empty levels with an underscore, handling NaN and 'Unnamed' values
        cleaned_levels = [str(level).strip() for level in col if pd.notna(level) and not str(level).startswith('Unnamed')]
        if not cleaned_levels:
            # If all levels are Unnamed or NaN, keep the original multi-index name or generate a default
            new_columns.append('_'.join(map(str, col)).replace('Unnamed: ', '').replace('level_0', '').replace('level_1', '').replace('level_2', '').strip('_').replace('__', '_'))
        else:
            new_columns.append('_'.join(cleaned_levels).replace('\n', ' ')) # Replace newline characters in column names
    return new_columns

df_new.columns = flatten_multiindex_columns(df_new)
# Drop the first row as it seems to be part of the header description and not actual data
df_new = df_new.iloc[1:].reset_index(drop=True)

# Rename the first two columns to be more descriptive based on typical data structure
df_new = df_new.rename(columns={'Men': 'Race_or_Ethnicity_Men', 'American Indian or Alaska Native': 'Race_or_Ethnicity_Women'})


# Display the cleaned DataFrame head and its info to confirm changes
display(df_new.head())
print(df_new.info())
```
