```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='observation_date', y='daily returns')
plt.title('Daily Returns of DJIA Over Time')
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.grid(True)
plt.xticks(rotation=45)=
plt.tight_layout()
plt.show()
```
