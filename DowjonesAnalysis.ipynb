```python
name = input("Please enter your name: ")

# Check if the name is empty
if not name.strip():
  print("Error: Name cannot be empty.")
# Check if the name is too long
elif len(name) >= 10:
  print ("Error: Input too long. Name should be less than 10 characters.")
# Check if the name contains only alphabetic characters (optional, depending on requirements)
elif not name.replace(' ', '').isalpha():
  print("Error: Name should only contain alphabetic characters.")
# Check if the name is all uppercase
elif name == name.upper():
  print ("Error: Bad Input. Name should not be all uppercase.")
else:
    # Display the name all in lowercase
    print(f"Valid name (lowercase): {name.lower()}")
```

```python
import pandas as pd # import pandas

# URL for the dataset
url = "https://www.ers.usda.gov/media/5628/vegetables-and-dry-pulses-trade-data.csv?v=74229"

df = pd.read_csv(url)
```

```python
display(df.head(10))

# Aggregation using Group By
df_trade = df.groupby (["Year", "Trade"])["Amount"].sum()
display(df_trade)
```

```python
# Filtering columns
df["UnitDesc"].unique()

# Let's filter to only get rows where the unit is thousand pounds
df_pounds = df[(df["UnitDesc"] == "Thousand pounds") & (df["Group"] == "Vegetables")]
df_pounds.groupby(["Year", "Trade"])["Amount"].sum().plot(kind="bar")
```
