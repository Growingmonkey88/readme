```python
import pandas as pd

url = "https://noaaocm.blob.core.windows.net/ais/csv2/csv2025/ais-2025-01-08.csv.zst"

df = pd.read_csv(url)
df
```

```python
df.index = pd.to_datetime(df.index)
df.index
```

```python
df.index = pd.to_datetime (df.index)
df.index
```

```python
df_cleaned = df.dropna (subset=["sog"])
display(df_cleaned.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Explicitly create a copy to avoid SettingWithCopyWarning
df_cleaned_copy = df_cleaned.copy()

# Extract the hour from the DateTimeIndex
df_cleaned_copy['hour'] = df_cleaned_copy.index.hour

# Count unique MMSI per hour
vessels_per_hour = df_cleaned_copy.groupby('hour')['mmsi'].nunique().reset_index()
vessels_per_hour.columns = ['hour', 'unique_vessels']

# Plot the data
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='unique_vessels', data=vessels_per_hour, marker='o')
plt.title('Number of Unique Vessels per Hour (UTC)')
plt.xlabel('Hour of Day (UTC)')
plt.ylabel('Number of Unique Vessels')
plt.xticks(range(0, 24)) # Ensure all hours are displayed on the x-axis
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Find the hour with the most and least boats
hour_most_boats = vessels_per_hour.loc[vessels_per_hour['unique_vessels'].idxmax()]
hour_least_boats = vessels_per_hour.loc[vessels_per_hour['unique_vessels'].idxmin()]

print(f"Hour with the most boats out: {hour_most_boats['hour']} UTC ({int(hour_most_boats['unique_vessels'])} unique vessels)")
print(f"Hour with the least boats out: {hour_least_boats['hour']} UTC ({int(hour_least_boats['unique_vessels'])} unique vessels)")
```

```python
# Calculate the duration when speed over ground (sog) is greater than 0
active_speed_intervals = henry_hudson_df[henry_hudson_df['sog'] > 0]

# Calculate the difference in time for each active speed interval
# This assumes that if sog > 0, the engine is active during that measurement period
# and we are interested in the duration between consecutive measurements.
# The first measurement with sog > 0 will not have a preceding time difference if it's the start of activity.

# Group consecutive rows where sog > 0 to find continuous active periods.
# A simple way to approximate total active time is to sum the time differences
# where sog is positive, or just count the number of positive sog readings
# and multiply by the average time interval, but a more accurate way involves grouping.

# Let's consider a simpler approach for now: count records where sog > 0 and calculate average time difference.
# For a more robust calculation, we'd need to identify continuous periods.

# For simplicity, let's first calculate the total number of records where sog > 0
num_active_records = len(active_speed_intervals)

# Calculate the time difference between consecutive records in the original df for average interval estimation
# Or, more directly, calculate the time difference for each record in the 'active_speed_intervals' where sog > 0
# However, this doesn't give us the *duration* directly, but rather the individual points.

# Let's refine the approach to calculate the actual duration of engine activity.
# We can consider the engine active if 'sog' is greater than 0.
# We need to calculate the time difference between consecutive 'sog > 0' entries.

# Filter for rows where sog is greater than 0
active_sog_df = henry_hudson_df[henry_hudson_df['sog'] > 0]

# Calculate the time difference between consecutive readings in seconds
time_diffs = (active_sog_df.index.to_series().diff()).dt.total_seconds()

# Sum the time differences to get total active time in seconds
total_active_seconds = time_diffs.sum()

# Convert to a more readable format (e.g., hours and minutes)
total_active_minutes = total_active_seconds / 60
total_active_hours = total_active_seconds / 3600

print(f"Total time Henry Hudson's engines were active (sog > 0): {total_active_seconds:.2f} seconds")
print(f"Total time Henry Hudson's engines were active (sog > 0): {total_active_minutes:.2f} minutes")
print(f"Total time Henry Hudson's engines were active (sog > 0): {total_active_hours:.2f} hours")

# Optionally, you might want to look at periods of actual acceleration (positive acceleration)
positive_acceleration_intervals = henry_hudson_df[henry_hudson_df['acceleration'] > 0]
num_positive_accelerations = len(positive_acceleration_intervals)

print(f"Number of instances with positive acceleration: {num_positive_accelerations}")
```

```python
def calculate_total_active_engine_time(mmsi_input):
    """
    Calculates the total active engine time for a given vessel (MMSI) for the day.

    Args:
        mmsi_input (int): The MMSI of the vessel.

    Returns:
        float: Total active engine time in hours.
    """
    vessel_df = df_cleaned[df_cleaned['mmsi'] == mmsi_input].sort_index()
    active_sog_df = vessel_df[vessel_df['sog'] > 0]

    if active_sog_df.empty:
        return 0.0

    time_diffs = (active_sog_df.index.to_series().diff()).dt.total_seconds()
    total_active_seconds = time_diffs.sum()
    total_active_hours = total_active_seconds / 3600

    return total_active_hours

# Example usage with Henry Hudson's MMSI
henry_hudson_mmsi = 366651000
total_active_hours_henry_hudson = calculate_total_active_engine_time(henry_hudson_mmsi)
print(f"Henry Hudson's total active engine time: {total_active_hours_henry_hudson:.2f} hours")
```
